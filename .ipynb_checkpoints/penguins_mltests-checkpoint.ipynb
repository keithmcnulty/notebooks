## Predicting Penguin Sex in the Palmer Penguins Dataset
This is not a big dataset, just for playing around.

```python
penguins = plp.load_penguins()
penguins.head()
```

```python
penguins.info()
```

```python
sns.pairplot(penguins, hue = "sex")
```

We see clear correlation between flipper length and body mass - not surprising.  The clusters seem to indcate that the different species are playing a role here, and that males and females might be reasonably predictable within their species.  Suggests that decision tree like models might be most successful at predicting the sex of a penguin from this sample.

## Logistic regression

```python
# lose any null values
penguins = penguins.dropna()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# create dummies for categorical data
dummies = pd.get_dummies(penguins[['species', 'island']], drop_first=True)
```

```python
dummies
```

```python
dummy_y = pd.get_dummies(penguins['sex'], drop_first=True)
dummy_y
```

```python
X = pd.concat([penguins[['bill_length_mm', 'bill_depth_mm', 
                         'flipper_length_mm', 'body_mass_g', 
                         'year']], dummies], 
              axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, dummy_y['male'], test_size = 0.3, random_state = 101)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
predicts = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predicts))
```

```python
print(classification_report(y_test, predicts))
```

Pretty decent accuracy here.  That will be hard to beat I think.

## Feature prep for other models

```python
## scale for subsequent models

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## K-Nearest Neighbours Classifier

```python
# use elbow method to find k
from sklearn.neighbors import KNeighborsClassifier

error_rate = []

for i in range(1, 40):
    modeli = KNeighborsClassifier(n_neighbors=i)
    modeli.fit(X_train, y_train)
    predicti = modeli.predict(X_test)
    error_rate.append(np.mean(predicti != y_test))
```

```python
error_rate = pd.DataFrame(dict(k=range(1,40), err=error_rate))
```

```python
sns.lineplot(data = error_rate, x = 'k', y = 'err')
```

```python
## k = 14 looks best

knnmodel = KNeighborsClassifier(n_neighbors=13)
```

```python
knnmodel.fit(X_train, y_train)
```

```python
knnpredicts = knnmodel.predict(X_test)
```

```python
print(confusion_matrix(y_test, knnpredicts))
```

```python
print(classification_report(y_test, knnpredicts))
```

Not as good as logit.

## Simple neural network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
annmodel = Sequential()

# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw


# input layer
annmodel.add(Dense(9,  activation='relu'))

# hidden layer
annmodel.add(Dense(3, activation='relu'))


# output layer
annmodel.add(Dense(units=1, activation='sigmoid'))

# Compile model
annmodel.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
annmodel.fit(x=X_train, 
          y=y_train, 
          epochs=25,
          validation_data=(X_test, y_test) 
          )
```

```python
annpredicts = annmodel.predict_classes(X_test)
```

```python
print(confusion_matrix(y_test, annpredicts))
```

```python
print(classification_report(y_test, annpredicts ))
```

Not as good.  There are a lot of parameters here - could try with dropouts and different activation functions.
